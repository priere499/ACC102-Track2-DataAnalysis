```python
#导入库
import wrds
import pandas as pd
```

```python
#连接 WRDS
db = wrds.Connection(wrds_username="yuxuanhuang")
```

```python
#获取苹果公司年度数据
sql_query = """
    SELECT tic, fyear, at, sale, ni
    FROM comp.funda
    WHERE tic = 'AAPL'
    AND fyear >= 2010
    ORDER BY fyear;
"""

# 执行查询
df = db.raw_sql(sql_query)

# 显示前5行
df.head()
```

```python
#基础查看
df.info()
df.describe()
```

```python
# 保存数据为 CSV 文件
df.to_csv("AAPL_WRDS_Financial_Data.csv", index=False)

print("CSV 文件已保存！")
```
